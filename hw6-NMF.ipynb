```python
file = open("vectors.txt","r")
data = file.readlines()
Nlines = len(data)
print("Nlines = ",Nlines)
```

```python
wfile = open("words_idx.txt","r")
words = wfile.readlines()
Nwords = len(words)
print("Nwords = ",Nwords)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ndocs = int(Nlines/2)
A = np.zeros((Nwords,Ndocs))
Doc_idx = np.zeros((Ndocs,))
label = np.zeros((Ndocs,))
for j in range(Ndocs):
    Doc_idx = data[2*j]
    line = data[2*j+1]
    line = line.split()
    numbers = [eval(i) for i in line]
    label = numbers[0]
    w_idx = np.array(numbers[1::2])-1 # make indices start from 0
    w_count = numbers[2::2]
    A[w_idx,j] = 1       
# np.savetxt("Amatrix.csv", A, delimiter=",")
```

```python
#-------------------------------part A----------------------------------
k = 10
W = np.random.rand(Nwords,k)
H = np.random.rand(k,Ndocs)
R = A - W@H
iter_max = 200
R_FroNorm = np.zeros(iter_max)

alpha=0.01
for i in range(iter_max):
    Wnew = W+alpha*(R @ np.transpose(H))
    Wnew = np.maximum(Wnew, 0)
    Hnew = H+alpha*(np.transpose(W) @ R)
    Hnew = np.maximum(Hnew, 0)
    W=Wnew
    H=Hnew
    R = A - W@H
    R_FroNorm[i] = np.linalg.norm(R,'fro')
for i in range(200):
    print (i, 'iter' , R_FroNorm[i])
```

```python
#-------------------------------part B----------------------------------
k = 10
W = np.random.rand(Nwords,k)
H = np.random.rand(k,Ndocs)
R = A - W@H

iter_max = 200
R_FroNorm = np.zeros(iter_max)

for itr in range(10): #number of iterations you want the algo to perform
    RH=R@np.transpose(H) 
    HH=H@np.transpose(H)
    
    WR=(np.transpose(W)@R)
    WW=(np.transpose(W)@W)
    
    for j in range(k): 
        
        u= np.zeros(Nwords)
        u = np.maximum(-W[:,j] , RH[:,j]/np.diag(HH)[j]) 

        #update W and R
        W[:,j] = W[:,j] + u
        R = R - np.outer(u,H[j,:])

        
        v= np.zeros(Nwords)
        v = np.maximum(-H[j,:] , WR[j,:] / np.diag(WW)[j]) #check diag. changed all i to j

        #update H and R
        H[j,:] = H[j,:] + v
        R = R - np.outer(W[:,j],v)
        
    #finished one pass through W and H (via columns of W and rows of H)
    R_FroNorm[itr] = np.linalg.norm(R,'fro')
    
R_FroNorm
```

```python
#-------------------------------part C----------------------------------
U, S, Vt = np.linalg.svd(A, full_matrices=True)

# Choose rank r 
r = 10
U_r = U[:, :r]           # First r columns of U
S_r = np.diag(S[:r]) # top r singular values
Vt_r = Vt[:r, :]          # First r rows of V^T
#print(U_r.shape,S_r.shape, Vh_r.shape)
A_r = U_r @ S_r @ Vt_r

Residual_FroNorm=np.linalg.norm(A-A_r,'fro')
Residual_FroNorm
```
